```python
import pandas as pd
import numpy as np

df = pd.read_csv('knn_data.csv')
X = df.drop('Natija', axis =1).values
y = df['Natija'].values

x_new = np.array([3,60])
```

```python
# 2-QADAM: euclidean() funksiyasini yozing

def euclidean(x1,x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
from collections import Counter

#3-QADAM: knn_predict() funksiyasini yaratish
def knn_predict(X,y,x_new,k):
    distances = []
    for i in range (len(X)):
        d = euclidean(X[i],x_new)
        
        #4-QADAM: Masofalarni hisoblab saqlash
        
        distances.append((d, y[i]))
    
    # 5-QADAM: Masofalar bo‘yicha saralash
    
    distances.sort(key = lambda x: x[0])
    
    #6-QADAM: Eng yaqin K ta qo‘shnini tanlash
    
    k_nearest = distances[:k]

    #🔹 7-QADAM: Ovoz berish (classification)
    
    votes = [label for dist, label in k_nearest]
    counts = Counter(votes)
    return counts.most_common(1)[0][0]

natija = knn_predict(X, y, x_new, k=3)
# 8-QADAM: To‘liq natijani qaytarish
if natija == 0:
    print("O'ta olmadi")
else: 
    print("O'tdi")
```
